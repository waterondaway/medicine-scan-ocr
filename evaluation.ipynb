```python
# Setting up Google Cloud Vision API
from google.cloud import vision
import os, io
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "senior-project-455216-772c73df2329.json"
client = vision.ImageAnnotatorClient()
```

```python
# Protocal for Evaluating OCR
# This code is used to evaluate the performance of an OCR system by comparing the OCR output with the ground truth text.
from jiwer import wer, cer

def normalize(text):
    return text.replace('\n', ' ').replace('  ', ' ').replace('   ', ' ').replace('    ', ' ').strip()
```

## Type Font - Regular Style

### 💬 Google Cloud Vision API

```python
count = 0
sum_wer_score = 0
sum_cer_score = 0
folder = ['ChakraPetch', 'Kanit', 'Krub', 'Sarabun', 'Prompt']
for i in folder:
    for j in os.listdir(f'picture/{i}-Augmentation/regular/'):
        count += 1
        print(f'Processing {j}...')
        file_path = f'picture/{i}-Augmentation/regular/{j}'
        with io.open(file_path, 'rb') as image_file:
            content = image_file.read()
        image = vision.Image(content=content)
        response = client.text_detection(image=image)
        texts = response.text_annotations
        ocr_text = texts[0].description 
        with open(f'ground_truth/ground_truth_{j.split("_")[2].split(".")[0]}.txt', 'r', encoding='utf-8') as f:
            ground_truth = f.read()


        wer_score = wer(normalize(ground_truth), normalize(ocr_text))
        sum_wer_score += wer_score
        cer_score = cer(normalize(ground_truth), normalize(ocr_text))
        sum_cer_score += cer_score

print(f'Average WER: {sum_wer_score/count}')
print(f'Average CER: {sum_cer_score/count}')
```

## Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

styles = ['Regular', 'Bold', 'Italic']

pytesseract = [0.55, 0.58, 0.56]
google_vision = [0.15, 0.20, 0.15]

x = np.arange(len(styles))
width = 0.35

fig, ax = plt.subplots(figsize=(8, 6))

bar1 = ax.bar(x - width/2, google_vision, width, label='Google Vision', color='#1f3b4d')
bar2 = ax.bar(x + width/2, pytesseract, width, label='Pytesseract', color='#b33939')

ax.set_ylabel('Word Error Rate (WER)')
ax.set_xlabel('Font Styles')
ax.set_title('Word Error Rate (WER) for Different OCR Models and Font Styles')
ax.set_xticks(x)
ax.set_xticklabels(styles)
ax.set_ylim(0, 1)
ax.legend()

for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
